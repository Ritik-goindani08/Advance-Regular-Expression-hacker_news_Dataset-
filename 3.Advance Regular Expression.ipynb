```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hk_news = pd.read_csv("Datasets/hacker_news.csv")
```

```python
title = hk_news["title"]
```

```python
# pattern = "sql","SQL","Sql"
```

```python
pattern = r"([Ss][Qq][Ll])"
tag = title[title.str.contains(pattern)].str.extract(pattern)
tag.head(40)
```

```python
# Mysql, Nosql
import re

pattern = r"(\w+sql)"
tag = title[title.str.contains(pattern, flags = re.I)].str.extract(pattern, flags = re.I)
tag.head()
```

```python
pattern = r"([Pp]ython\s[\d.]+)"
tag = title[title.str.contains(pattern, flags = re.I)].str.extract(pattern, flags = re.I)
tag.head()
```

```python
pattern = r"(\b[Cc]\b[^.+])"
tag = title[title.str.contains(pattern, flags = re.I)]
tag
```

# lookaround
1. lookback
2. lookforward

1. lookforward zzz(?=abc), zzz(?!abc)
2. lookback zzz(?<=abc), zzz(?<!abc)

```python
pattern = r"(?<!Series\s)\b[Cc]\b(?!.+)"
tag = title[title.str.contains(pattern, flags = re.I)]
tag
```

```python
pattern = r"\b(\w+)\s\1\b"
tag = title[title.str.contains(pattern, flags = re.I)]
tag
```
